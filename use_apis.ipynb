```python
# the others are for debugging but can be used stand-alone to see how different parts of code work
API_URL = "http://localhost:8002/simulate_marketing"
# API_URL = "http://localhost:8001/create_persona"
# API_URL = "http://localhost:8001/retrieve_personas"

# Payload for simulating a marketing campain
payload = {
    "age_range": "40-70",
    "region": "Eindhoven",
    "gender": "Any",
    "n": 6,
    "marketing_text": """
**Meet the NoteLoop™ – The Last Notebook You'll Ever Need.**  
Write, scan, erase, repeat. The NoteLoop™ combines the timeless feel of pen on paper with modern cloud syncing. Save your notes to the cloud with a snap, wipe the page clean, and start fresh – all in one sleek, eco-friendly package. Smarter notes, zero waste.
"""} # Used chatGPT for this marketing text, but I think you can see what the point is

response = requests.post(API_URL, json=payload)

# check if there is a valid response
if response.status_code == 200:
    data = response.json()
```

```python
df = pd.DataFrame(data['marketing_results'])
```
